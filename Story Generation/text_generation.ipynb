```python
import transformers
from transformers import AutoModelForCausalLM
from transformers import AutoTokenizer
from transformers import BitsAndBytesConfig
import torch
import os
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
model_id = "bigscience/bloom-7b1"
```

```python
bnb_config = BitsAndBytesConfig(
    load_in_8bit=True
)
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```python
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config=bnb_config,
    dtype=torch.float16,
    device_map="auto"
)
```

```python
def generate_story(
    prompt,
    model,
    tokenizer,
    temprature = 0.5,
    max_length = 500
):
    inputs = tokenizer(prompt, return_tensors = "pt").to(model.device)
    
    outputs = model.generate(
        inputs["input_ids"],
        max_new_tokens = max_length,
        do_sample = True,
        temperature = temprature,
        top_p = 0.9,
        repetition_penalty = 1.15,
        no_repeat_ngram_size=3,
    )
    response = tokenizer.decode(
        outputs[0, inputs["input_ids"].shape[-1]:],
        skip_special_tokens = True
    )
    
    print(response)
```

```python
prompt = """Once upon a time,there was a young child exploring the forest.
            Suddenly, they discovered ancient ruins hidden beneath the overgrown vines.
            The child stepped inside and"""
generate_story(prompt, model, tokenizer)
```

```python
prompt_2 = """Once upon a time,there was an abandoned house accross a big river.
            it was the biggest challenge among the kids of the village.
            Two sisters decided to uncover it's secrets by """

generate_story(prompt_2, model, tokenizer)
```

```python
prompt_3 = """Once upon a time, A rabit, a bear and a mouse decided to go on a great journey. 
            After walking for a while they found a map and  """
generate_story(prompt_3, model, tokenizer)
```
